## 2015, September

```python
table2015_1 = soup_doc.find_all('tbody')[0]
table2015_1
```

```python
print(table2015_1.text)
```

```python
rows = table2015_1.find_all('tr')
rows
```

```python
table2015_1 = soup_doc.find_all('tbody')[0]

rows_2015_1 = table2015_1.find_all('tr')

rows=[]

for row in rows_2015_1[1:]:
    cells_2015_1 = row.find_all('td')
    #print(cells_2015_1)
    cell = {}
    
    party = cells_2015_1[0]
    cell['party'] = party.text
    
    leader = cells_2015_1[1]
    cell['leader'] = leader.text
    
    seats = cells_2015_1[2]
    cell['seats'] = seats.text
    
    percentage = cells_2015_1[3]
    cell['percentage'] = percentage.text
    
    votes = cells_2015_1[4]
    cell['votes'] = votes.text
    
    print(cell)
    rows.append(cell)
    
```

```python
df_2015_1 = pd.DataFrame(rows)
df_2015_1
```

```python
df_2015_1['date'] = '2015, Sept'
```

```python
df_2015_1['percentage'] = df_2015_1['percentage'].str.replace(',','.')
```

```python
df_2015_1['percentage'] = df_2015_1['percentage'].astype(float)
```

```python
df_2015_1['votes'] = df_2015_1['votes'].str.replace('.','')
```

```python
df_2015_1['votes'] = df_2015_1['votes'].astype(int)
```

```python
df_2015_1
```

```python
df_2015_1_en['party'] = df_2015_1_en['party'].replace({"Συνασπισμός Ριζοσπαστικής Αριστεράς":'SYRIZA',
                                                      'Νέα Δημοκρατία':'New Democracy',
                                                      'Λαϊκός Σύνδεσμος Χρυσή Αυγή':'Golden Dawn',
                                                      'Δημοκρατική Συμπαράταξη (ΠΑ.ΣΟ.Κ. - ΔΗΜ.ΑΡ.)':'Democratic Alignment',
                                                      'Κομμουνιστικό Κόμμα Ελλάδας':'Communist Party',
                                                      'Το Ποτάμι':'Potami',
                                                      'Ανεξάρτητοι Έλληνες - Εθνική\r\n\t\t\tΠατριωτική Δημοκρατική Συμμαχία':'ANEL',
                                                      'Ένωση Κεντρώων':'Union of Centrists'})
```
