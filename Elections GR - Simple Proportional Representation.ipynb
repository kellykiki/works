```python
df1 = pd.read_csv('elections.csv')
```

```python
results2019 = pd.read_csv('results2019.csv')
```

```python
df1.head()
```

```python
df1 = df1.drop('Unnamed: 0',axis=1)
```

```python
df1.head()
```

```python
results2019
```

```python
df1 = pd.concat([df1,results2019])
df1.head()
```

```python
df1 = df1.drop('Unnamed: 0',axis=1)
df1.head()
```

```python
df2 = pd.DataFrame(data=df1.groupby('date').votes.sum())
df2 = df2.reset_index()
df2 = df2.rename(columns={'votes':'total_votes'})
df2
```

```python
df3 = pd.merge(df1,df2,on='date')
```

```python
df3.head()
```

```python
df3['law_2016'] = round(df3.votes*300/df3.total_votes)
df3.head()
```

```python
df5 = pd.DataFrame(data=df3.groupby('date').seats.max()).reset_index()
df5
```

```python
df7 = pd.DataFrame(data=df3.groupby('date').law_2016.max()).reset_index()
df7
```

```python
df = pd.merge(df5,df7,on='date')
df
```

```python
df.dtypes
```

```python
df['date'] = df['date'].replace({"1989, Jun":'1989.06',
                                 "1989, Nov": '1989.11',
                                 "2012, Jun": '2012.06',
                                 "2012, May": '2012.05',
                                 "2015, Jan": '2015.01',
                                 "2015, Sept": '2015.09'})
```

```python
df['date'] = df.date.astype(float)
df.dtypes
```

```python
df = df.sort_values(by=['date'])
df
```

```python
ax = df.plot(kind='line',x='date',figsize=(15,6),marker='o')

ax_thres = plt.axhline(y=151, color='grey',linewidth=2, linestyle=':',alpha = 1)

# ax.set_xticks([1974,1977,1981,1985,1989,1990,1991,1993,1996,2000,2004,2007,2009,2012,2013,2015,2016])
# ax.set_xticklabels(['1974','1977','1981','1985','1989','1990','1991','1993','1996','2000','2004','2007','2009','2012','2013','2015','2016'])

ax.set_yticks([50,100,150,200,250])
ax.set_yticklabels(['50','100','150','200','250'])
ax.set_xlim(1973,2020)
# ax.set_ylim(50,250)

# plt.savefig('historical.pdf')
plt.show()
```
